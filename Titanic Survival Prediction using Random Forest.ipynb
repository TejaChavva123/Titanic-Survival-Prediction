**Dividing the data into train and test data**

```python
from sklearn.model_selection import train_test_split
```

```python
x = df.iloc[:,1:]
y = df['Survived']
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state = 57)
```

**Model building build using Random Forest Classifier**

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators = 5)
model.fit(x_train,y_train)
```

```python
y_pred_test = model.predict(x_test)
```

```python
from sklearn.metrics import classification_report,accuracy_score
```

```python
print("Testing accuracy is: ",accuracy_score(y_test,y_pred_test))
```

**Model building build using Decision Forest Classifier**

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
model.fit(x_train,y_train)
```

```python
y_pred_test = model.predict(x_test)
print("Testing accuracy is: ",accuracy_score(y_test,y_pred_test))
```

**From here u can note that testing accuracy is more in the Random Forest Classifier than Decision Tree Classifier**

**Variance will be low after ensemble Technique**
